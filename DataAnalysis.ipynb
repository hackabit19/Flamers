```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
import random
```

```python
base_path = r"C:\Users\FIRDOUS SULTANA\Downloads\annotations\\"
```

```python
def analysis(movie_name):
    file_dir = base_path + "\\" + movie_name
    files = [i for i in os.listdir(base_path) if os.path.isfile(os.path.join(base_path,i)) and \
         movie_name in i]
    for file in files:
        concept = file.split('_')[1]
        concept = concept[:len(concept)-4]
        df = pd.read_csv(base_path + file, sep = " ",header = None, names = ['start_frame','end_frame', 'label'])
        if(df.empty):
            continue
        df['label'].fillna(concept,inplace =True)
        #df['label'].value_counts().plot(kind='bar')
        palettes = ['Set1','Set2','Set3']
        ax = sns.countplot(df['label'],palette = random.choice(palettes))
        ax.tick_params(axis='x', rotation=70)
        plt.title(concept)
        plt.xlabel('Concept Detail')
        plt.show()
```

```python
analysis('Armageddon')
```

```python
analysis('FightClub')
```
